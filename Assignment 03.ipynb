```python
import os
import re
from random import shuffle
```

```python
current_dir = os.getcwd()
current_dir
```

```python
def remove_html(doc):
    cleaner = re.compile('<.*?>')
    cleaned_text = re.sub(cleaner, '', doc)
    return cleaned_text
```

```python
def remove_special_characters(doc):
    removed_text = re.sub(r"\W+|_", " ", doc)
    return removed_text
```

```python
doc_list = []

doc_dir_neg = os.chdir('D:\\IRWA Assignment\\aclImdb\\test\\neg')
file_list_neg = os.listdir(doc_dir_neg)
for file in file_list_neg:
    with open(file, 'r', encoding = 'utf-8') as f:
        text = f.read()
        html_remove_text = remove_html(text)
        sp_chr_removed_text = remove_special_characters(html_remove_text).lower()
        doc_list.append((sp_chr_removed_text.split(), 'neg'))
        
        
doc_dir_pos = os.chdir('D:\\IRWA Assignment\\aclImdb\\test\\pos')
file_list_pos = os.listdir(doc_dir_pos)
for file in file_list_pos:
    with open(file, 'r', encoding = 'utf-8') as f:
        text = f.read()
        html_remove_text = remove_html(text)
        sp_chr_removed_text = remove_special_characters(html_remove_text).lower()
        doc_list.append((sp_chr_removed_text.split(), 'pos'))
```
